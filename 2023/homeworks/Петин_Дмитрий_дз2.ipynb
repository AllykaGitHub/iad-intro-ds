**Задание 12. Выполните задание 11 с помощью циклов. Сравните время выполнения.**

```python
unique_eth = df['ethnicity'].unique()
unique_edu= df['education'].unique()
```

```python
%time

# task 1
task1 = []
for eth in unique_eth:
  tmp_sum = 0
  count = 0
  for i in range(len(df)):
    if df.loc[i, 'ethnicity'] == eth:
      tmp_sum += df.loc[i, 'reading score']
      count += 1
  task1.append(round(tmp_sum / count, 3))


# task 2
task2 = []
for edu in unique_edu:
  min = 101
  for i in range(len(df)):
    if df.loc[i, 'education'] == edu and df.loc[i, 'writing score'] < min:
      min = df.loc[i, 'writing score']
  task2.append(min)


print('ethnicity')
for i, eth in enumerate(unique_eth):
  print(eth, ' ', task1[i])
print()
print('education')
for i, edu in enumerate(unique_edu):
  print(f'{edu:<20} {task2[i]}')

%timeit
#метод груп бай работает быстрей и не такой громосткий 
```
